```python
plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots(figsize=(8, 4))

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

## Typical line plot

```python
import matplotlib.pyplot as plt

month = ['Jan','Feb','Mar']
actual_sales = [37560, 34200, 35860 ]
forecast_sales = [38600, 34500, 34900]

plt.subplots(figsize=(8, 4))
plt.plot(month, actual_sales, label='Actual Sales')
plt.plot(month, forecast_sales, '--o', label='Forecast Sales', color='r')
plt.ylabel('sales in month')
plt.xlabel('2023 Q1')
plt.grid(True)
plt.legend()
plt.title('Sales Forecast vs Actual Sales as of 2023')
```

```python
plt.subplots(figsize=(8, 4))

vaccine = ['1st dose', '2nd dose', '3rd dose', '4th dose']
counts = [6904243, 6779247, 5764953, 675118]
bar_colors = ['tab:blue', 'tab:green', 'tab:orange', 'tab:red']

plt.bar(vaccine, counts, label=vaccine, color=bar_colors)

plt.ylabel('Population with Doses Administered')
plt.title('Vaccine Doses Administered as of Dec 2022')

plt.show()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
ax.set_ylim(-1.5, 1.5)
plt.show()
```

```python
np.random.seed(19680801)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots(figsize=(8, 6))

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--', color='orange')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.grid()
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()

plt.show()
```
